```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img = io.imread('../input/cnn-data-sources/yo.png')
img_gray = color.rgb2gray(img)
```

```python
print('La imagen original es: ', img.shape)
print('La imagen en escala de grises es: ', img_gray.shape)
```

```python
plt.imshow(img_gray)
```

```python
kernel = np.array([[-1,0,1], [-1,0,1], [-1,0,1]])
```

```python
imagen_convulucionada = nd.convolve(img_gray, kernel)
```

```python
plt.imshow(imagen_convulucionada)
```

```python
fig, axes = plt.subplots(1,2,figsize = (15,10))
axes[0].imshow(img_gray, cmap = plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(imagen_convulucionada, cmap = plt.cm.gray)
axes[1].set_title('Convulucionada')
axes[1].axis('off')
```
